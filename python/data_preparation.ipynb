```python
import pandas as pd
```

```python
#Read both sheets
df1=pd.read_excel("online_retail_II.xlsx", sheet_name="Year 2009-2010")
df2=pd.read_excel("online_retail_II.xlsx", sheet_name="Year 2010-2011")
```

```python
#Combine
df=pd.concat([df1, df2], ignore_index=True)
print(df.head())
```

```python
#info
print(df.info())
```

```python
#summary stats
print(df.describe())
```

```python
#check NA
print(df.isnull().sum())
```

```python
#check duplicates
print("Duplicate rows:", df.duplicated().sum())
```

```python
#quality check
print("Negative quantity:", (df["Quantity"] < 0).sum())
print("Zero or negative price:", (df["Price"] <= 0).sum())
```

```python
#cancelled item
cancelled=df["Invoice"].astype(str).str.startswith("C").sum()
print("Cancelled invoices:", cancelled)
```

```python
#NA in ID and description
print("Missing Customer ID:", df["Customer ID"].isnull().sum())
print("Missing Description:", df["Description"].isnull().sum())
```

```python
#save as tsv file for easier pig integration
df.to_csv("online_retail_ii.tsv", sep="\t", index=False, header=False)
```
